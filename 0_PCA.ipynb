```python
import pandas as pd

# df_titanic = pd.read_csv('./data/titanic.csv')
df_titanic = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/PCA_analysis_guide/data/titanic.csv')
df_titanic.head()
```

```python
# @title embarked vs pclass

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df_titanic['embarked'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df_titanic, x='pclass', y='embarked', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
plt.show()
```
